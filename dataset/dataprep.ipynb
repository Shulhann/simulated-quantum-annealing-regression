```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

#### DATASET 1 : California Housing Data
source : https://www.kaggle.com/datasets/camnugent/california-housing-prices

```python
dataset_1 = pd.read_csv('raw dataset/housing.csv')
dataset_1.describe()
```

```python
numerical_cols = dataset_1.select_dtypes(include=['float64', 'int64']).columns
dataset_1[numerical_cols] = MinMaxScaler().fit_transform(dataset_1[numerical_cols])
dataset_1 = dataset_1[numerical_cols]
```

```python
dataset_1 = dataset_1.dropna()
```

```python
dataset_1.to_csv('processed dataset/dataset_1.csv', index=False)
```

#### DATASET 4 : Auto MPG
source : https://archive.ics.uci.edu/dataset/9/auto+mpg

```python
# Read the data (assuming it's saved as 'auto-mpg.data')
dataset_4 = pd.read_csv(
    'raw dataset/auto-mpg.data',
    delim_whitespace=True,  # splits on any whitespace
    quotechar='"',          # handles quoted car names
    na_values='?',          # treats '?' as NaN
    header=None,            # no header row in the file
    names=[                 # manually assign column names
        'mpg', 'cylinders', 'displacement', 'horsepower',
        'weight', 'acceleration', 'model_year', 'origin', 'car_name'
    ]
)

dataset_4.describe()
```

```python
numerical_cols = dataset_4.select_dtypes(include=['float64', 'int64']).columns
dataset_4[numerical_cols] = MinMaxScaler().fit_transform(dataset_4[numerical_cols])
dataset_4 = dataset_4[numerical_cols]
```

```python
dataset_4 = dataset_4.dropna()
```

```python
new_column_order = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'mpg']
dataset_4 = dataset_4[new_column_order]
```

```python
dataset_4.to_csv('processed dataset/dataset_4.csv', index=False)
```
